```python
# Revised IDS training pipeline (balance for PortScan, BruteForce, etc.)
# Requirements:
# pip install pandas scikit-learn xgboost imbalanced-learn joblib matplotlib seaborn

import os
import json
import time
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE, RandomOverSampler
import xgboost as xgb

# -----------------------
# User config
# -----------------------
file_path = "cleaned_ids_dataset.csv"   # <-- update path if needed
label_col = "label"
min_samples_for_smote = 8   # if class count < this, merge into "Other"
random_state = 42
test_size = 0.30

# Output dir
out_dir = "models"
os.makedirs(out_dir, exist_ok=True)

# -----------------------
# 1) Load dataset
# -----------------------
df = pd.read_csv(file_path)
print("Loaded:", df.shape)
print("Columns:", df.columns.tolist())

# -----------------------
# 2) Basic label counts and merging extremely rare classes into 'Other'
# -----------------------
if label_col not in df.columns:
    raise ValueError(f"Label column '{label_col}' not found in dataset")

counts = df[label_col].value_counts().sort_values(ascending=False)
print("\nOriginal label counts (top 30):")
print(counts.head(30))

# Merge classes with extremely low support into 'Other'
rare_classes = counts[counts < min_samples_for_smote].index.tolist()
if rare_classes:
    print("\nMerging very rare classes into 'Other':", rare_classes)
    df[label_col] = df[label_col].apply(lambda x: 'Other' if x in rare_classes else x)
else:
    print("\nNo classes needed merging.")

# Recompute counts
counts = df[label_col].value_counts().sort_values(ascending=False)
print("\nLabel counts after merge (top 50):")
print(counts.head(50))

# -----------------------
# 3) Features & Target
# -----------------------
X = df.drop(columns=[label_col])
y = df[label_col].astype(str)   # ensure string

# If there are categorical features, encode them (simple LabelEncoder)
# NOTE: If you used numeric-only features already, this will be minimal.
cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()
encoders = {}
if cat_cols:
    print("\nEncoding categorical columns:", cat_cols)
    for c in cat_cols:
        le = LabelEncoder()
        X[c] = le.fit_transform(X[c].astype(str))
        encoders[c] = le

# 4) Encode label
le_label = LabelEncoder()
y_enc = le_label.fit_transform(y)
label_mapping = dict(zip(le_label.classes_, range(len(le_label.classes_))))
print("\nLabel mapping (class -> encoded):")
print(label_mapping)

# -----------------------
# 5) Train/test split (stratified)
# -----------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc, test_size=test_size, random_state=random_state, stratify=y_enc
)
print("\nTrain/Test shapes:", X_train.shape, X_test.shape)

# -----------------------
# 6) Prepare SMOTE sampling_strategy
#    - target_count = median class count in training set
#    - for classes below target_count we will request upsampling
# -----------------------
train_counts = pd.Series(y_train).map({v:k for k,v in le_label.transform(le_label.classes_).items()}) if False else None
# Use counts on training labels
unique, counts_arr = np.unique(y_train, return_counts=True)
train_label_counts = dict(zip(unique, counts_arr))
# compute median count (but not > max)
median_count = int(np.median(list(train_label_counts.values())))
median_count = max(median_count, 50)  # ensure a reasonable floor (tunable)
print("\nTraining label counts (encoded -> count):")
print(train_label_counts)
print("Median target count for oversampling:", median_count)

# Build sampling_strategy dict for SMOTE: for classes with count < median_count
sampling_strategy = {}
for cls, cnt in train_label_counts.items():
    if cnt < median_count:
        sampling_strategy[int(cls)] = median_count

if not sampling_strategy:
    print("\nNo SMOTE needed (classes already >= median).")
else:
    print("\nSMOTE sampling_strategy sample (encoded_class: target_count):")
    print(sampling_strategy)

# -----------------------
# 7) Apply SMOTE / RandomOverSampler safely
# -----------------------
X_res, y_res = X_train.copy(), y_train.copy()
if sampling_strategy:
    try:
        # choose k_neighbors based on smallest class in sampling_strategy (must be >=1)
        smallest = min([np.sum(y_train==c) for c in sampling_strategy.keys()])
        k = min(5, max(1, smallest-1))
        sm = SMOTE(sampling_strategy=sampling_strategy, random_state=random_state, k_neighbors=k, n_jobs=-1)
        X_res, y_res = sm.fit_resample(X_train, y_train)
        print("\nAfter SMOTE resample shape:", X_res.shape, np.unique(y_res, return_counts=True))
    except Exception as e:
        print("SMOTE failed:", e)
        print("Falling back to RandomOverSampler (safe for very small classes).")
        ros = RandomOverSampler(sampling_strategy=sampling_strategy, random_state=random_state)
        X_res, y_res = ros.fit_resample(X_train, y_train)
        print("\nAfter ROS resample shape:", X_res.shape, np.unique(y_res, return_counts=True))
else:
    print("\nNo oversampling performed; using original training set.")

# -----------------------
# 8) Train XGBoost classifier
# -----------------------
num_classes = len(le_label.classes_)
model = xgb.XGBClassifier(
    objective="multi:softprob",
    num_class=num_classes,
    n_estimators=200,
    learning_rate=0.1,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    tree_method="hist",
    random_state=random_state,
    use_label_encoder=False,
    eval_metric="mlogloss"
)

print("\nTraining XGBoost...")
model.fit(X_res, y_res)

# -----------------------
# 9) Evaluate on test set
# -----------------------
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("\nðŸ”¹ Accuracy:", acc)

print("\nðŸ”¹ Classification Report:")
print(classification_report(y_test, y_pred, target_names=le_label.inverse_transform(sorted(set(y_test)))))

# Full confusion matrix (with label order)
labels_order = le_label.classes_
cm = confusion_matrix(y_test, y_pred, labels=range(len(labels_order)))
print("\nðŸ”¹ Confusion Matrix (rows=true, cols=pred):")
print(cm)

# Plot confusion matrix (large multiclass)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=False, cmap="Blues")
plt.title("Confusion Matrix (encoded labels)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Feature importance (XGBoost)
try:
    importances = model.feature_importances_
    feat_imp = pd.Series(importances, index=X.columns).sort_values(ascending=False).head(20)
    plt.figure(figsize=(8,6))
    sns.barplot(x=feat_imp.values, y=feat_imp.index)
    plt.title("Top Features (XGBoost)")
    plt.show()
except Exception as e:
    print("Could not plot feature importance:", e)

# -----------------------
# 10) Save model + encoders + feature list + metadata
# -----------------------
ts = time.strftime("%Y%m%d_%H%M%S")
model_path = os.path.join(out_dir, f"xgb_ids_{ts}.pkl")
joblib.dump(model, model_path)
joblib.dump(le_label, os.path.join(out_dir, f"label_encoder_{ts}.pkl"))
joblib.dump(encoders, os.path.join(out_dir, f"feature_encoders_{ts}.pkl"))
with open(os.path.join(out_dir, f"features_{ts}.json"), "w") as f:
    json.dump(X.columns.tolist(), f)

print("\nSaved model and artifacts under:", out_dir)
print("Model file:", model_path)
```
